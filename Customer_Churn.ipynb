#converted 0 and 1 into yes and no of seniorcitizen to understand easy

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create crosstab (counts of SeniorCitizen vs Churn)
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Convert to percentage of total (row-wise)
ct_perc = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar
ax = ct_perc.plot(kind="bar", stacked=True, figsize=(4,5), colormap="Set2")

# Add percentage labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("Churn by SeniorCitizen(Stacked Bar Chart)")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen")
plt.legend(title="Churn",bbox_to_anchor=(0.9,0.9))       #customize legend location , we can  also use loc='upper right'
plt.show()
```

```python

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define rows & cols for subplots (3x3 grid for 9 plots)
n = len(cols)
rows = 3
cols_per_row = 3

fig, axes = plt.subplots(rows, cols_per_row, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, hue="Churn", ax=axes[i])
    axes[i].set_title(f"Churn by {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=30)

# Remove empty subplot if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
